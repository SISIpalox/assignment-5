```python
import numpy as np

def makeSpiral(Max):
    N = 1
    while(N**2 < Max):
        N += 1
    dirs = [(0, 1), (-1, 0), (0, -1), (1, 0)]

    spiral = np.zeros((N, N), dtype=int)

    if N%2 == 1:
        coordx = int((N-1)/2)
        coordy = int((N-1)/2)
    else:
        coordx = int(N/2)
        coordy = int(N/2)-1
    
    now = np.array([coordx, coordy])
    dirs_arg = 0
    max_arrows = 1
    count_arrows = 0
    reach_max = 0

    for i in range(1, Max+1):
        spiral[tuple(now)] = i
        if count_arrows == max_arrows:
            dirs_arg = (dirs_arg + 1) % 4
            reach_max += 1
            count_arrows = 0
            
        if reach_max == 2:
            max_arrows += 1
            reach_max = 0

        now += dirs[dirs_arg]
        count_arrows += 1

    return spiral

# Generating the spiral matrix with Max set to 529 (corrected from 500 to match the function call)
spiral_matrix_529 = makeSpiral(529)
spiral_matrix_529
```

```python
import matplotlib.pyplot as plt

# Create a color map based on the condition described
color_map = np.zeros_like(spiral_matrix_529, dtype='object')

# Assign colors based on the modulo of 4
color_map[spiral_matrix_529 % 4 == 0] = 'red'   # Modulo 4 equals 0
color_map[spiral_matrix_529 % 4 == 1] = 'blue'  # Modulo 4 equals 1
color_map[spiral_matrix_529 % 4 == 2] = 'green' # Modulo 4 equals 2
color_map[spiral_matrix_529 % 4 == 3] = 'yellow'# Modulo 4 equals 3

# Plotting
fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(np.zeros_like(spiral_matrix_529), aspect='equal', interpolation='none', cmap='gray', extent=[0, color_map.shape[1], 0, color_map.shape[0]])

# Overlay the colors
for i in range(color_map.shape[0]):
    for j in range(color_map.shape[1]):
        ax.text(j + 0.5, color_map.shape[0] - i - 0.5, spiral_matrix_529[i, j], 
                ha='center', va='center', color=color_map[i, j])

# Adjusting the appearance of the plot
ax.set_xlim(0, color_map.shape[1])
ax.set_ylim(0, color_map.shape[0])
ax.set_xticks([])
ax.set_yticks([])
plt.axis('off')
plt.show()
```
